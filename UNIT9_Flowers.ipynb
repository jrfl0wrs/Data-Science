````python
import os
import nltk
import urllib.request
from bs4 import BeautifulSoup
import requests
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
nltk.download('punkt')
URL = 'http://www.analytictech.com/mb021/mlk.htm'
````

# Read And Clean Content from URL

---

````python
# Page content from Website URL
page = requests.get(URL)

# Function to remove tags
def remove_tags(html):

    # parse html content
    soup = BeautifulSoup(html, "html.parser")

    for data in soup(['style', 'script']):
        # Remove tags
        data.decompose()

    # return data by retrieving the tag content
    return ' '.join(soup.stripped_strings)


# Print the extracted data
web_text = remove_tags(page.content)
````

#Create MLK File

---



````python
f = open("mlk.txt", "w")
f.write(web_text)
f.close()
````

# Create Corpus from web Content

---



````python
corpus = PlaintextCorpusReader(os.getcwd(),"mlk.txt")
````

# Word Count and Common Words:

````python
mlk_words = corpus.words()

# WORD COUNT
print("The Number of words in this corpus:", len(mlk_words))
# TOP 5 WORDS
freq_words = nltk.FreqDist(mlk_words)
print("Top 5 Words in this Document:",freq_words.most_common(5))
````

# TOKEN LIST:

````python
base_file = open("mlk.txt","rt")
raw_text = base_file.read()
base_file.close()
token_list = nltk.word_tokenize(raw_text)

#Remove Punct

punkt_remove = list(filter(lambda token: nltk.tokenize.punkt.PunktToken(token).is_non_punct, token_list))
print("Total Tokens after removing punctuation:", len(punkt_remove))
````

#Switch To Lowercase

````python
# Lower Case
lower_case = [word.lower() for word in punkt_remove]
print("Tokens after Lowercase and remove punctuation:", lower_case[:5])
````

WORD COUNT AFTER CLEAN:

````python
f = open("cleaned_mlk.txt", "a")
for word in lower_case:
  f.write(str(word) + " ")
f.close()
````

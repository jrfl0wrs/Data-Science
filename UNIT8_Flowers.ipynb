````python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
````

# LINEAR REGRESSION USING PYTHON AND SKLEARN



> **extract data from your data frame into a numpy array:**


>*HINT | .reshape()*


````python
#IMPORT DATA
df=pd.read_excel('Stroke.xlsx')
#REPLACE Y/N WITH 1/0
df['Smoker'] = df['Smoker'].replace({'Yes':1,'No':0})

# EXTRACT INTO NUMPY ARRAYS
y = df['Risk'].values
x_age = df['Age'].values
x_sstatus = df['Smoker'].values
x_bloodpressure = df['BloodPressure'].values

#Reshape NumpyArrays to fit LinReg
y = y.reshape(1,-1)
x_age = x_age.reshape(1,-1)
x_bloodpressure = x_bloodpressure.reshape(1,-1)
x_sstatus = x_sstatus.reshape(1,-1)

#call linear Regression
LinReg = LinearRegression()
df.head()
````

# AGE VS RISK POINT PLOT



```
Age V.S. Risk:
```



````python
# Generate a point plot showing the risk factor and the stroke risk.
print(x_age)
plt.plot(x_age,y,'r^')
plt.xlabel("AGE")
plt.ylabel("RISK")
plt.show()
````

````python
# Using the format "y = mx + b", show the regression equation.
# Show the R2 value.
LinReg.fit(x_age,y)
print(LinReg.intercept_,LinReg.coef_)
````

#BLOOD PRESSURE VS RISK POINT PLOT



```
Blood-Pressure V.S. Risk:
```



````python
# Generate a point plot showing the risk factor and the stroke risk.
print(x_bloodpressure)
plt.plot(x_bloodpressure,y,'r^')
plt.xlabel("Blood Pressure")
plt.ylabel("RISK")
plt.show()
````

````python
# Using the format "y = mx + b", show the regression equation.
# Show the R2 value.
LinReg = LinearRegression()
LinReg.fit(x_bloodpressure,y)
print(LinReg.intercept_,LinReg.coef_)
````

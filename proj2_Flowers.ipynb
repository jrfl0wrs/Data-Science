```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```



# **Data from the Centers for Disease Control (CDC)**

```python
#live data from the Centers for Disease Control (CDC)
# Build a data frame the attached file:covid2020StateDate.csv

df = pd.read_csv("covid2020StateDate.csv")
```

```python
# new data frame with only the following two columns:  date and newpos
new_frame = df.drop(['state'], axis=1)
print(new_frame.head())
```

# **Two Column Dataframe: date and newpos (New Positive Cases)**

```python
# Convert the date column to a Python date column.
new_frame['date'] = pd.to_datetime(new_frame['date'],format="%Y%m%d")
print(new_frame)
```

```python
# Combine rows so that there is just one row for each date
date_sums = new_frame.groupby("date")["newpos"].sum()
print(date_sums)
```

```python
x_pos = np.arange(len(date_sums))
# Create a bar graph with the date as the x-axis and the number of cases as the y-axis,
# which appropriate titles for the graph and axes.
# Create bars
# plt.bar(x_pos, new_frame.groupby("date")["newpos"].sum())
plot_frame = new_frame.groupby("date")["newpos"].sum()
plot_frame.sort_index().plot(kind='bar')
# Show graphic
plt.xlabel("DATE")
plt.ylabel("# OF CASES")
plt.show()
```

# **BOX PLOT: SEPAL LENGTH**

```
Box plot || Boxplot
Is a type of chart often used in explanatory data analysis.
Box plots visually show the distribution of numerical data and
skewness by displaying the data quartiles (or percentiles) and averages.
```


````python
#Generate a box plot diagram for the sepal length, grouped by the species.
sns.boxplot( x=df["species"], y=df["sepal_length"] )
````

````python
# Explain the meaning of: 1 - 4
````


# **Explain the meaning of:**


1) the top and bottom lines of the box


> The tope line of the box is the upper quartile  also known as the third quartile. 25% of data is above this value, 75% being below. The Bottom line of the box is considered the lower quartile, only 25% of the data fall below this line.


2) the line in the middle of the box


> The line in the middle of the box marks the mid-point, also known as, the median. The middle point of the data and is shown by the line that divides the box into two parts


3) the top and bottom "whiskers"


> The upper and lower whiskers represent scores outside the middle 50%, the lower being the bottom 25% of scores and the upper being the top 25% of scores.


4) the diamond point underneath the bottom whisker of the virginica species.


> The diamond point represents a outlier. The outlier is numerically distant from the rest of the data,located outside the whiskers of the box plot.

# **SCATTER PLOT: SEPAL LENGTH AND WIDTH**

```
SCATTERPLOT || SCATTER PLOT
A scatter plot is a graphical display that shows the relationships or associations
between two numerical variables (or co-variables), which are represented as points
(or dots) for each pair of scores.

A scatter plot indicates the strength and direction of the correlation between the co-variables.
```


````python
# Generate a scatter plot.
sns.lmplot( x="sepal_width", y="sepal_length", data=df, fit_reg=False, hue='species', legend=False)
# Move the legend to an upper part of the plot
plt.legend('upper right')
plt.show()
````
